##### Just combining the final dataset

```python
import pandas as pd

a = pd.read_csv("../CT-2022/2022_merged_data.csv")
b = pd.read_csv("../CT-2023/2023_merged_data.csv")
c = pd.read_csv("../CT-2021/2021_merged_data.csv")
d = pd.read_csv("../CT-2024/2024_merged_data.csv")

df = pd.concat([a, b, c, d], ignore_index=True)
```

```python
test = df.isna().sum()
for col, count in test.items():
    if count != 0:
        print(col, count)
```

```python
df.info()
```

```python
df['total_officers'] = df['male_officer'] + df['female_officer']

df = df.drop(columns=[
    'male_civilian', 'male_officer+male_civilian', 
    'female_civilian', 'female_officer+female_civilian', 
    'agency_id'
])

df = df[(df['population'] != 0) & (~df['population'].isna())]

df = df.rename(columns={
    'agency_name': 'city',
    'incident_date': 'date',
    'incident_hour': 'hour'
})
```

```python
# Date Features
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['dayofweek'] = df['date'].dt.dayofweek  # 0 = Monday, 6 = Sunday
```

```python
order = [
    'year', 'month', 'day', 'hour', 'date', 'dayofweek',
    'city', 'population', 
    'male_officer', 'female_officer', 'total_officers', 
    'offense_name', 'offense_category_name', 'location_area'
]

df = df[order]
```

```python
df.to_csv("combined_data.csv", index=False)
```
